```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
ride_data_to_load = pd.read_csv("Resources/ride_data.csv")
city_data_to_load = pd.read_csv("Resources/city_data.csv")


# Read the City and Ride Data
city_data = pd.DataFrame(city_data_to_load)
ride_data = pd.DataFrame(ride_data_to_load)
# Combine the data into a single dataset
pyber_df = pd.merge(
    ride_data, city_data,how="outer", on="city",right_index=True).sort_index()
# Display the data table for preview

pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_group = pyber_df.loc[pyber_df['type'] == 'Urban']
rural_group = pyber_df.loc[pyber_df['type'] == 'Rural']
suburban_group = pyber_df.loc[pyber_df['type'] == 'Suburban']

urban_cities=urban_group.groupby(['type',"city"]).mean()
urban_ride_count=urban_group.groupby(['type',"city"]).count()
urban_drivers=urban_cities['driver_count']

suburban_cities=suburban_group.groupby(['type',"city"]).mean()
suburban_ride_count=suburban_group.groupby(['type',"city"]).count()
suburban_drivers=suburban_cities['driver_count']

rural_cities=rural_group.groupby(['type',"city"]).mean()
rural_ride_count=rural_group.groupby(['type',"city"]).count()
rural_drivers=rural_cities['driver_count']
#ride_count


# Build the scatter plots for each city types


plt.scatter(urban_ride_count['driver_count'],urban_cities['fare'],s=urban_drivers*10,color="peru",alpha=.75,linewidths=.8,edgecolor='k',label="Urban")
plt.scatter(suburban_ride_count['driver_count'],suburban_cities['fare'],s=suburban_drivers*10,color="lightskyblue",alpha=.75,linewidths=1,edgecolor='k',label="Suburban")
plt.scatter(rural_ride_count['driver_count'],rural_cities['fare'],s=rural_drivers*10,color="gold",alpha=.75,linewidths=1,edgecolor='k',label="Rural")
# Incorporate the other graph properties

plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
lgnd=plt.legend(title='City Types',markerscale=.75,scatterpoints=1)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
plt.grid(True)
plt.annotate("     Note:",(40,36))
plt.annotate("     Circle size correlates with driver count per city.",(40,34.5))

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure[save as image;code is in one of the activities]
plt.savefig("Images/bubble_plot.png")
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_total_fare=urban_group['fare'].sum()
suburban_total_fare=suburban_group['fare'].sum()
rural_total_fare=rural_group['fare'].sum()
total_fare=[urban_total_fare,rural_total_fare,suburban_total_fare]
# Build Pie Chart
plt.pie(total_fare,explode=(.1,0,0),labels=['Urban','Rural','Suburban'],autopct='%1.1f%%',
        shadow=True, startangle=280, colors = ["lightcoral", "gold", "lightskyblue"])
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("Images/fare_by_city_type.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_group_rides=len(urban_group)
suburban_group_rides=len(suburban_group)
rural_group_rides=len(rural_group)
total_rides=[urban_group_rides,rural_group_rides,suburban_group_rides]
# Build Pie Chart
plt.pie(total_rides,explode=(.1,0,0),labels=['Urban','Rural','Suburban'],autopct='%1.1f%%',
        shadow=True, startangle=265, colors = ["lightcoral", "gold", "lightskyblue"])
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("Images/rides_by_city_type.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivers=[urban_drivers.sum(),rural_drivers.sum(),suburban_drivers.sum()]
# Build Pie Charts
plt.pie(total_drivers,explode=(.1,0,0),labels=['Urban','Rural','Suburban'],autopct='%1.1f%%',
        shadow=True, startangle=230, colors = ["lightcoral", "gold", "lightskyblue"])
plt.title("% of Total Drivers by City Type")
#Save Figure
plt.savefig("Images/drivers_by_city_type.png")
```

```python
# Show Figure
plt.show()
```
